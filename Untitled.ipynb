```python
import pandas as pd
df = pd.read_csv('ligaesp.csv')
print(df[df.columns.tolist()])
```

```python
print(df['Team 1'])
```

```python
conteo_total = df['Team 1'].nunique() ##cuenta el total de equipos sin repetir
print("\nNúmero total de elementos en la columna 'Team 1':", conteo_total)
```

```python
# Renombrar una columna
df.rename(columns={'FT': 'Resultados'}, inplace=True)
print(df)
goles_por_equipo = df.groupby('Team 1')['Goles'].sum()

# Mostrar el total de goles por equipo
print("Total de goles por equipo (sin repeticiones en 'Team 1'):")
print(goles_por_equipo)
```

```python
# Calcular el total de goles en cada partido sumando los goles de 'Team 1' y 'Team 2'
df['Total de goles'] = df['Team 1'] + df['Team 2']

# Encontrar el partido con más goles
partido_mas_goles = df.loc[df['Total de goles'].idxmax()]

# Mostrar el partido con más goles
print("Partido con más goles:")
print(partido_mas_goles)
```

```python

# Filtrar el DataFrame por la ronda 2
round_2 = df[df['Round'] == 2]

# Obtener una lista de equipos únicos en la ronda 2
equipos_round_2 = pd.concat([round_2['Team 1'], round_2['Team 2']]).unique()

# Crear un diccionario para almacenar la suma de goles de cada equipo en la ronda 2
goles_por_equipo_round_2 = {}

# Iterar sobre los equipos y calcular la suma de los goles para cada uno en la ronda 2
for equipo in equipos_round_2:
    # Filtrar los resultados donde el equipo aparece en 'Team 1' o 'Team 2' en la ronda 2
    resultados_equipo = round_2[(round_2['Team 1'] == equipo) | (round_2['Team 2'] == equipo)]
    # Sumar los goles del equipo en los partidos de la ronda 2 donde participa
    goles_equipo = resultados_equipo['Resultados'].sum()
    # Almacenar la suma de goles en el diccionario
    goles_por_equipo_round_2[equipo] = goles_equipo

# Encontrar el equipo que ha marcado más goles en la ronda 2
equipo_mas_goleador_round_2 = max(goles_por_equipo_round_2, key=goles_por_equipo_round_2.get)
goles_mas_goleador_round_2 = goles_por_equipo_round_2[equipo_mas_goleador_round_2]

# Imprimir el equipo más goleador y la cantidad de goles en la ronda 2
print("En la ronda 2, el equipo más goleador fue:", equipo_mas_goleador_round_2)
print("Número total de goles marcados por el equipo:", goles_mas_goleador_round_2)
```

```python
# Filtrar el DataFrame para obtener todas las filas donde el Atlético de Madrid esté presente en 'Team 1' o 'Team 2'
partidos_atletico = df[(df['Team 1'] == 'Atlético Madrid') | (df['Team 2'] == 'Atlético Madrid')]

# Seleccionar las columnas relevantes (round, resultado y rival)
partidos_atletico_filtrados = partidos_atletico[['Round', 'Resultados', 'Team 1', 'Team 2']]

# Imprimir los partidos del Atlético de Madrid
print("Partidos del Atlético de Madrid:")
print(partidos_atletico_filtrados)
```

```python
import mysql.connector
import matplotlib.pyplot as plt

# Conectar a la base de datos
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="root",
    database="resultados"
)

# Crear un cursor
cursor = conn.cursor()

# Realizar una consulta para obtener la puntuación de los equipos
cursor.execute("SELECT puntuacion FROM resultados")

# Obtener los resultados de la consulta
puntuaciones = cursor.fetchall()

# Cerrar la conexión
conn.close()

# Obtener fluctuación acumulada
fluctuacion_acumulada = []
puntos_acumulados = 0

# Calcular la fluctuación acumulada de puntuaciones
for puntuacion in puntuaciones:
    if puntuacion[0] is not None:
        puntos_acumulados += puntuacion[0]
        fluctuacion_acumulada.append(puntos_acumulados)
    else:
        # Manejar el caso cuando la puntuación es None
        print("Advertencia: Se encontró un valor de puntuación None.")

# Crear el gráfico de línea
plt.figure(figsize=(15, 10))
plt.plot(range(1, len(fluctuacion_acumulada) + 1), fluctuacion_acumulada, marker='o', linestyle='-')
plt.xlabel('Partidos')
plt.xticks(range(1, 40))  # Esto generará una secuencia de números del 1 al 40.
plt.ylabel('Puntos acumulados')
plt.title('Fluctuación acumulada de resultados del Atlético de Madrid')
plt.grid(True)

# Agregar puntos acumulados en cada marcador
for i, txt in enumerate(fluctuacion_acumulada):
    plt.annotate(txt, (i+1, fluctuacion_acumulada[i]), textcoords="offset points", xytext=(0,10), ha='center')

# Agregar puntos acumulados en cada marcador
for i, txt in enumerate(fluctuacion_acumulada):
    plt.annotate(txt, (i+1, fluctuacion_acumulada[i]), textcoords="offset points", xytext=(0,10), ha='center')

# Mostrar el gráfico
plt.show()

# Contar la cantidad de victorias, derrotas y empates
victorias = fluctuacion_acumulada.count(3)
derrotas = fluctuacion_acumulada.count(0)
empates = fluctuacion_acumulada.count(1)

# Mostrar el recuento de victorias, derrotas y empates
print(victorias, "victorias")
print(derrotas, "derrotas")
print(empates, "empates")
```
